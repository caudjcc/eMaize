```python
def get_table_(loc):
    cc = np.where(np.array(['BJ',  'HeB','HN', 'JL', 'LN']) ==loc)[0][0]
    table_touse =phenotypes[phenotypes['loc'] ==loc]
    parent_table = table_touse.pivot_table(values='number', index=['m'], columns=['f'], dropna=False)
    male_ids = [male_name[i] for i in range(0, parent_table.shape[0])]
    female_ids = [female_name[i] for i in range(0, parent_table.shape[1] )]
    parent_table = parent_table.loc[male_ids, female_ids]
    parent_table = parent_table - 6210*cc
    a = {}
    for i in range(3):
        trait = ['DTT','PH','EW'][i]
        a[i] = np.take(np.ravel(table_touse[trait].values), parent_table)
    return a[0],a[1],a[2]
```

```python
value_intab = {}
location = ['BJ',  'HeB','HN', 'JL', 'LN']
trait = ['DTT','PH','EW']
for i in range(5):
    value_intab[location[i]] = {}
    for j in range(3):
        value_intab[location[i]][trait[j]] =get_table_(location[i])
```

```python
def diffrenet_table(loc,loc_):
    a = {}
    b = {}
    c = {}
    a[0],a[1],a[2] = get_table_(loc)
    b[0],b[1],b[2] = get_table_(loc_)
    plt.style.use('ggplot')
    fig, ax = subplots(3,1, figsize=(20, 12))
    for i in range(3):
        trait = ['DTT','PH','EW'][i]
        ax[i].matshow(a[i]-b[i], cmap=cm.RdBu_r)
        ax[i].set_title('Differences of %s'%trait + ' in %s'%loc +' and %s' %loc_,fontsize=15)
        ttl = ax[i].title
        ttl.set_position([.5, 1.15])
        c[i] = a[i]-b[i]
    return c[0],c[1],c[2]
```

```python
diffrenet_tab = {}
count =0
for i in range(4):
    for j in np.arange(i+1,5):
        diffrenet_tab[count] = diffrenet_table(location[i],location[j])
        count +=1
```
