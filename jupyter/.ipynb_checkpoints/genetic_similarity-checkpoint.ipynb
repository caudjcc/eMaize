```python
import h5py
```

```python
import os
eigvals = {}
for n_snps in os.listdir('output/gsm'):
    n_snps = int(n_snps)
    eigvals[n_snps] = []
    for i in os.listdir('output/gsm/%d'%n_snps):
        i = int(i)
        with h5py.File('output/gsm/%d/%d'%(n_snps, i)) as f:
            eigvals[n_snps].append(f['eigvals'][:])
```

```python
from scipy.stats import pearsonr
# compare eigenvalues within the same SNP set size
r_within = {}
for n_snps in eigvals.keys():
    r_within[n_snps] = []
    n_runs = len(eigvals[n_snps])
    for i in range(n_runs - 1):
        for j in range(i + 1, n_runs):
            r_within[n_snps].append(pearsonr(eigvals[n_snps][i][:235],
                                             eigvals[n_snps][j][:235])[0])
# compare eigenvalues between different sizes of SNPs
r_between = []
n_classes = len(eigvals)
classes = eigvals.keys()
for i in range(n_classes - 1):
    for j in range(i + 1, n_classes):
        r_between.append(pearsonr(eigvals[classes[i]][0][:235],
                                  eigvals[classes[j]][0][:235])[0])
```

```python
import pandas as pd
def generate_parent_table(phenotype_file):
    phenotypes = pd.read_table(phenotype_file)
    pedigree = phenotypes['pedigree'].str.split('_', expand=True)
    pedigree.columns = ['f', 'X', 'm']
    phenotypes = pd.concat([phenotypes, pedigree], axis=1)
    phenotypes['number'] = np.arange(phenotypes.shape[0])
    parent_table = phenotypes.pivot_table(values='number', index=['m'], columns=['f'], dropna=False)
    male_ids = ['m%d' % i for i in range(1, parent_table.shape[0] + 1)]
    female_ids = ['f%d' % i for i in range(1, parent_table.shape[1] + 1)]
    parent_table = parent_table.loc[male_ids, female_ids]
    return parent_table
phenotype_file = 'emaize_data/phenotype/pheno_emaize.txt'
parent_table = generate_parent_table(phenotype_file)
phenotypes = pd.read_table(phenotype_file)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
```

```python
X_avail = V[:, :237][(phenotypes['type'] == 'training').values]
y_avail = phenotypes['trait1'][(phenotypes['type'] == 'training').values].values
X_test = V[:, :237][(phenotypes['type'] == 'test').values]
y_test = phenotypes['trait1'][(phenotypes['type'] == 'test').values].values
index_avail = np.nonzero((phenotypes['type'] == 'training').values)
index_test = np.nonzero((phenotypes['type'] == 'test').values)
```

```python
X_avail.shape, y_avail.shape
```

```python
from tqdm import tqdm_notebook
```

```python
# cross-validation using linear regression on eigenvectors
kfold = KFold(10, shuffle=True)
metrics = {'r2': [], 'pcc': []}
pbar = tqdm_notebook(total=10)
for train_index, valid_index in kfold.split(X_avail):
    X_train, X_valid = X_avail[train_index], X_avail[valid_index]
    y_train, y_valid = y_avail[train_index], y_avail[valid_index]
    model = Ridge(alpha=0.01)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    metrics['r2'].append(r2_score(y_valid, y_pred))
    metrics['pcc'].append(pearsonr(y_valid, y_pred)[0])
    pbar.update(1)
pbar.close()
```

```python
# transform the phenotypes using eigenvectors (approximately average phenotype values of the same parent)
y_transformed_k = np.clip(np.dot(V1.T, y_avail), -4, 4)
fig, ax = plt.subplots(figsize=(16, 3))
ax.plot(y_transformed_k)
```

```python
# calculate the average phenotypes
phenotypes_parent = np.take(phenotypes['trait1'].values, parent_table)
phenotypes_female = np.nanmean(phenotypes_parent, axis=0)
phenotypes_male = np.nanmean(phenotypes_parent, axis=1)
phenotypes_parent = np.nan_to_num(np.concatenate([phenotypes_male, phenotypes_female]))
fig, ax = plt.subplots(figsize=(16, 3))
ax.plot(phenotypes_parent)
```

```python
# try to map transformed phenotype values to average phenotypes for each parent using linear combination
x = (phenotypes_parent - phenotypes_parent.mean())/phenotypes_parent.std()
y = (y_transformed_k - y_transformed_k.mean())/y_transformed_k.std()
xxT = np.dot(x[:, np.newaxis], x[np.newaxis, :]) + np.identity(237)*0.01
yxT = np.dot(y[:, np.newaxis], x[np.newaxis, :])
A = np.dot(yxT, np.linalg.inv(xxT))
# set 10 largest values to 0
A[:, np.argsort(-np.abs(A.mean(axis=0)))[:10]] = 0
print(pearsonr(y, np.dot(A, x)))
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
ax[0].scatter(y, np.dot(A, x), s=3)
ax[1].matshow(A, cmap=cm.RdBu)
```
